1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
# На вход данной функции поступает строка с описанием образования
def get_education(education):
    # Метод split() разбивает строку на слова по пробелу - получаем список слов в строке и заносим его в переменную edu_word_list.
    edu_word_list = education.split(' ')
    if edu_word_list[0] == "Высшее": education_type = 'высшее'
    if edu_word_list[0] == "Неоконченное": education_type = 'неоконченное высшее'
    if edu_word_list[0] == 'Среднее' and edu_word_list[1] != 'специальное': education_type = 'среднее'
    if edu_word_list[0] == 'Среднее' and edu_word_list[1] == 'специальное': education_type = 'среднее специальное'
    return education_type

hh_data['Образование'] = hh_data['Образование и ВУЗ'].apply(get_education)
display(hh_data['Образование'].value_counts())
```

Как вы понимаете, нам необходимо выделить каждый параметр в отдельный столбец.

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

Гарантируется, что структура одинакова для всех строк в таблице. Вы можете воспользоваться этим.*

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
import re

"""Функция, которая возвращает опыт работы в месяцах в формате float"""
def get_experience(exp):
    # Значения для пропусков и скрытых пропусков
    if exp is np.NaN or exp == 'Не указано': return np.NaN

    experience = re.findall(r'\d+', exp)                 # Список чисел
    experience_words = re.findall(r'(?i)[а-я]+', exp)    # Список слов
    
    # Если опыт исчисляется только в месяцах
    if experience_words[2] == 'месяц' or experience_words[2] == 'месяца' or experience_words[2] == 'месяцев':
        return float(int(experience[0]))
    # Если опыт исчисляется в годах и месяцах
    elif experience_words[3] == 'месяц' or experience_words[3] == 'месяца' or experience_words[3] == 'месяцев':
        return float(int(experience[0]) * 12 + int(experience[1]))
    # Если опыт исчисляется только в годах
    else:
        return float(int(experience[0]) * 12)
            
    
hh_data['Опыт работы (месяц)'] = hh_data['Опыт работы'].apply(get_experience)
display(hh_data['Опыт работы (месяц)'].median())
hh_data = hh_data.drop('Опыт работы', axis=1)             # Удаление признака опыт работы из таблицы
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# Списки с категориями занятости и графика
occupations = ['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка']
schedules = ['полный день', 'сменный график', 'гибкий график', 'удаленная работа', 'вахтовый метод']

# Цикл, который создает новые признаки с названиями категорий, значения которых — результат проверки принадлежности категории к перечню категорий в ячейке таблицы.
for occupation, schedule in zip(occupations,schedules):
    hh_data[occupation] = hh_data['Занятость'].apply(lambda x: occupation in x)
    hh_data[schedule] = hh_data['График'].apply(lambda x: schedule in x)
    
# Удаление признаков "Занятость" и "График"
hh_data = hh_data.drop('Занятость', axis=1)
hh_data = hh_data.drop('График', axis=1)

# Сколько людей ищут проектную работу И волонтёрство (в обоих столбцах стоит True)?
print(hh_data[hh_data['проектная работа'] & hh_data['волонтерство']].shape[0], 'человек ищут проектную работу И волонтёрство')
# Сколько людей хотят работать вахтовым методом И с гибким графиком (в обоих столбцах стоит True)?
print(hh_data[hh_data['вахтовый метод'] & hh_data['гибкий график']].shape[0], 'человек хотят работать вахтовым методом И с гибким графиком')
```

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# Новый DataFrame из полученного файла с данными курсов валют
exchange_data = pd.read_csv('data/ExchangeRates.csv', sep=',')

# Очистка таблицы exchange_data от лишних столбцов
exchange_data.drop(columns=['per', 'time', 'vol'], axis=1, inplace=True)
# Переименование столбцов в exchange_data 
exchange_data = exchange_data.rename(columns={
    'currency':'Валюта', 
    'date':'Дата', 
    'close':'Цена закрытия',
    'proportion':'Пропорция'
    })

# Переименование столбца в hh_data
hh_data = hh_data.rename(columns={'Обновление резюме':'Дата'}) 

# Переводим признак "Дата" (hh_data) в формат datetime и извлекаем из признака дату
hh_data['Дата'] = pd.to_datetime(hh_data['Дата'])
# Переводим признак "Дата" (exchange_data) в формат datetime
exchange_data['Дата'] = pd.to_datetime(exchange_data['Дата'])

# Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
"""Функция, которая возвращает сумму ( без указания валюты)"""
def get_amount(zp):
    zp_ls = zp.split()
    if zp_ls[1] == 'грн.' or zp_ls[1] == 'KGS': amount = int(zp_ls[0])
    elif zp_ls[1] == 'KZT': amount = int(zp_ls[0])
    elif zp_ls[1] == 'сум': amount = int(zp_ls[0])
    else: amount = int(zp_ls[0])
    return amount    
    
"""Функция, которая возвращает название валюты в формате iso"""
def get_currency(zp):
    zp_ls = zp.split()
    if zp_ls[1] == 'грн.': return 'UAH'
    elif zp_ls[1] == 'бел.руб.': return 'BYN'
    elif zp_ls[1] == 'сум': return 'UZS'
    elif zp_ls[1] == 'руб.': return 'RUB'
    else: return zp_ls[1]

# Преобразования в таблице hh_data    
hh_data['Зарплата'] = hh_data['ЗП'].apply(get_amount)  # Сумма желаемой заработной платы 
hh_data['Валюта'] = hh_data['ЗП'].apply(get_currency)  # Валюта
hh_data = hh_data.drop('ЗП', axis=1)                   # Удаление признака 'ЗП'

# Объединение талиц
hh_merged = hh_data.merge( exchange_data,
    left_on=['Дата', 'Валюта'],
    right_on=['Дата','Валюта'],
    how='left'
    )

# Создание списка дат (без времени) из таблицы hh_merged
hh_dates = hh_merged['Дата'].tolist()
hh_dates = [str(hh_dates[i]).split(' ')[0] for i in range(len(hh_merged))]

# Создание списка дат (без времени) из таблицы exchange_data
exch_dates = exchange_data['Дата'].tolist()
exch_dates = [str(exch_dates[i]).split(' ')[0] for i in range(len(exchange_data))]

# Вспомогательные списки
close = []
prop = []
zp = []
        
for i in range(0, len(hh_merged)):
    # Если зарплата в рублях, тогда пропорция и цена закрытия = 1
    if hh_merged['Валюта'][i] == 'RUB':
        prop.append(1)
        close.append(1)
        zp.append(hh_merged['Зарплата'][i])
        continue
    for j in range(0, len(exchange_data)):
        # Для валюты: если совпадают даты и валюты
        if hh_dates[i] == exch_dates[j] and hh_merged['Валюта'][i] == exchange_data['Валюта'][j]:
            close.append(exchange_data['Цена закрытия'][j])
            prop.append(exchange_data['Пропорция'][j])
            z=float(hh_merged['Зарплата'][i]) * float(exchange_data['Цена закрытия'][j]) / float(exchange_data['Пропорция'][j])
            zp.append(z)
            
# Заполняем признаки данными полученных списков
hh_merged['Цена закрытия'] = close
hh_merged['Пропорция'] = prop
hh_merged['ЗП (руб)'] = zp

# Возвращаем столбец 'Дата' обратно к формату datetime
hh_merged['Дата'] = pd.to_datetime(hh_merged['Дата'])

#Чему равна желаемая медианная заработная плата соискателей в нашей таблице (в рублях)?
print('Желаемая медианная зарплата составляет:', hh_merged['ЗП (руб)'].median())
```

Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
"""Функция, которая ищет выбросы методом z-отклонений"""
def outliers_z_score_mod(data, feature, left=3, right=3, log_scale=False):
    if log_scale:
        x = np.log(data[feature] + 1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = (mu - left*sigma)
    upper_bound = (mu + right*sigma)
    outliers = data[(x < lower_bound)|(x > upper_bound)]
    cleaned = data[(x > lower_bound)&(x < upper_bound)]
    return outliers, cleaned

outliers, cleaned = outliers_z_score_mod(hh_merged, 'Возраст', right=4, log_scale=True )
print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
print(f'Результирующее число записей: {cleaned.shape[0]}')


"""Построение гистограммы 'Распределения возраста в логарифмическом масштабе'"""
fig, ax = plt.subplots(1,1, figsize=(8,4))
log_hh =np.log(hh_merged['Возраст'] + 1)
histplot = sns.histplot(log_hh, bins=30, ax=ax)
histplot.axvline(log_hh.mean(), color='k', lw=2)
histplot.axvline(log_hh.mean() + 4*log_hh.std(), color='k', ls='--', lw=2)
histplot.axvline(log_hh.mean() - 3*log_hh.std(), color='k', ls='--', lw=2)
histplot.set_title('Распределение возраста в логарифмическом масштабе');

# Аномальные значения возраста?
# print(outliers['Возраст'])
# Ответы: 15, 15, 100

hh_merged= hh_merged[hh_merged['Возраст'] > 15]   # Сохраняем датафрейм с возрастом соискателей более 15 лет
hh_merged= hh_merged[hh_merged['Возраст'] < 100]  # Сохраняем датафрейм с возрастом соискателей менее 100 лет
```
